```python
import asyncio
from artemis.models.metar import OnsMetarApiSuite
from artemis.core.http_engine import Request
from artemis.utils import read_json
from artemis.models.location_model import Location
from matplotlib.cbook import flatten
import urllib
```

```python
file = read_json("resources/sul.json")
cities_list = []
for city in file['cities']:
    my_location = Location()
    my_location.set_location(city)
    cities_list.append(my_location)

id_locale_list = []
[id_locale_list.append(location.icao_id) for location in cities_list]
id_locale_list = list(flatten(id_locale_list))
    
model = OnsMetarApiSuite(id_locale_list=[id_locale_list[0]])
url = model.base_url +"dataInicioPeriodo={}&dataFimPeriodo={}&estacaoId={}&mnemonico=DRYT&modelo=METAR&pagina=1&itensPorPagina=500"
model.set_request_dict_by_date_range(date_begin="2021-10-05 03:00:00", date_end="2021-10-06 19:00:00", url=url)
getter = Request(data_dict=model.get_data_dict())
await getter.get(method='async')
getter.data_dict = getter.get_data_dict()    
```
